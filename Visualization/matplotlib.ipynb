```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating a figure and axes
fig, ax = plt.subplots(figsize=(5,3))

# setting the background colour
fig.set_facecolor('lightgrey')
ax.set_facecolor('yellow')

#plotting the data
x = [5,7,9,12,85]
y = [8,6,58,21,10]
ax.plot(x,y,
        marker = 'o',
        markersize= 5,
        markerfacecolor = 'red',
        markeredgecolor = 'black',
        linewidth = 2,
        linestyle ='--',
        color = 'green',
        alpha = 0.8)

#setting the tittle and labels
ax.set_title('Line Plot', fontsize=10,color='red')
ax.set_xlabel('X-axis', fontsize=8, color='blue')
ax.set_ylabel('Y-axis', fontsize=8, color='blue')

ax.axhline(y=30,color='r',linestyle='--',linewidth=1)
ax.axvline(x=45, color='r',linestyle='--',linewidth=1)

ax.axhspan(30,50, color='black',alpha=0.5)
ax.axvspan(10,45, color='black', alpha=0.5)

ax.text(10,45, 'Line Plot', fontsize=10, color='red')
```
